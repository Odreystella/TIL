```python
# sungnam.csv -> sungnam.json
# 1. sungnam.csv을 'r'로 불러온다.(csv.reader(파일객체)사용)
# 2. column name가 붙어 있기 때문에 colunm name을 dict의 key값으로 사용하기 위한 작업을 함
# 3. 만든 데이터를 json에 'w'하기 (json.dump)
# 4. 만든 json 파일을 'r'하기(json.load)

import csv 
import json


with open('./sungnam.csv', 'r', encoding='EUC-KR') as f:  # csv파일로 읽기
    csved = csv.reader(f)
    user_list = [row for row in csved]
#user_list

data={'users':[]}                                  # csv 파일안의 데이터를 {}타입으로 만들어주기
for row in user_list[1:]:
    dong_dict={}
    for index, key in enumerate(user_list[0]):
        dong_dict[key] = row[index]
    data['users'].append(dong_dict)   # 하나의 동이 dong_dict에 만들어지면, 'user'의 value값에 append
        
with open('./sungnam.json', 'w', encoding='EUC-KR') as g: # 만든 데이터를 json으로 쓰기
    json.dump(data,g, ensure_ascii=False)
    

with open('./sungnam.json', 'r', encoding='EUC-KR') as f: # json으로 불러오기
    des_data = json.load(f)
    for row in des_data['users']:
        print(row) 
    
```
