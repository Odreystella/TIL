```python
# 1. 코로나 라이브 웹사이트
import requests


url = 'https://apiv2.corona-live.com/updates.json?timestamp=1613547340934'
response = requests.get(url)
response.status_code
```

```python
result_dict = response.json()
```

```python
result_dict
```

```python
for row in result_dict:
    print("'일시': {} 확진자 {}건 발생, '알림메시지': {}".format(row['datetime'], row['cases'], row['src']))
```

```python
# 2. 코로나 위키피디아 사망, 회복, 확진 케이스
from bs4 import BeautifulSoup
import lxml

url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_in_South_Korea'
response = requests.get(url)
response
```

```python
html_text = response.text
soup = BeautifulSoup(html_text, 'lxml')

div_barbox = soup.find('div', {'class':'barbox'})
last_15 = div_barbox.find_all('tr',{'id':'mw-customcollapsible-2021feb-l15'})
date = last_15[0].find('td', {'class':'bb-04em'})
data = last_15[0].find_all('span',{'class':'cbs-ibr'})
cases = data[0].text
deaths = data[1].text
print(cases, deaths)
data
```

```python
for item in last_15:
    print('Date: {} / cases:{} / deaths: {}'.format(item.find('td', {'class':'bb-04em'}).text, 
                                                    item.find_all('span',{'class':'cbs-ibr'})[0].text,
                                                    item.find_all('span',{'class':'cbs-ibr'})[1].text))
    
```
