```python
# Arbitarary Arguments(arguments에 선택권이 있음, 사용자가 여러개 입력할 수 있음)
# argument를 여러개 받고 싶을 떄, parameter에 *를 붙임

def print_(*value):  # def print_(*args)가 관습적인 표현
    print(value, type(value))
    for i in value:
        print(i, end=" ")

print_(1,2,3,4)
```

```python
print_(1,2,3,4)
print('\n')
print(1,2,3,4)
```

```python
# Keyword Arguments
# key와 value로 parameter를 지정해서 값을 넣어줄 수 있음, 원하는 만큼

def get_user_info(user_name, real_name, **kwargs):
    print(user_name)
    print(real_name)
    print(kwargs, type(kwargs))
```

```python
get_user_info('fckim', 
              'Fastcampus Kin', 
              phone='01012345684', 
              locale='Seoul, KR')
```

```python
def get_user_info(**kwargs):
    print(kwargs, type(kwargs))
    for k,v in kwargs.items():
        print('{}:{}'.format(k,v))
    print('Your phone_num is {}'.format(kwargs['phone']))
```

```python
# domain name('www.google.com' naver daum)
# path admin.html
# query string
# ex) complete_uri('www.google.com',
#         utm_source='social',
#         utm_medium='campaign')
# https://www.google.com/admin.html?utm_source=social&utm_medium=campaign
```

```python
def complete_uri(domain_name,protocol='http://',path='admin.html',**query):
    query_string = ""    
    for k,v in query.items():
        query_string += '{}={}&'.format(k,v)
    query_string = query_string[:-1]
    print('{}{}/{}?{}'.format(protocol, domain_name, path, query_string, end=""))
```

```python
complete_uri('www.google.com',
         utm_source='social',
         utm_medium='campaign')
```

```python
complete_uri('www.google.com')
```

```python
# 조건문 사용하는 경우
def complete_uri(domain_name, protocol='http', **query):
    url = '{}://{}/admin.html'.format(protocol, domain_name)
    if not query:
        print(url)
    else:
        query_string = "?"
        for k,v in query.items():
            query_string += '{}={}&'.format(k,v)
        query_string = query_string[:-1]
        return url + query_string

complete_uri('www.google.com', 
             'sftp',
             utm_source='social',
             utm_medium='campaign')
```

```python
# leap_year comprehension
# leap_year: 4의 배수이면서, 100의 배수가 아니거나, 400의 배수인 연도
# 2000 = 윤년
# 2100 = 평년
years = [year for year in range(1999, 2101+1)]
is_leap_year = ['leap_year' if year%4==0 and (year%100!=0 or year%400==0) else 
                year for year in years]
is_leap_year
```

```python
# is_leap_year을 함수로 만들기
years = [year for year in range(1999, 2101+1)]
def is_leap(year):
    if year%4==0 and (year%100!=0 or year%400==0):
        return 'leap_year'
    else:
        return year
    
is_leap_year = [is_leap(year) for year in years]
```
